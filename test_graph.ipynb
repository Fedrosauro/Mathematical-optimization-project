```python
import networkx as nx
import matplotlib.pyplot as plt

# Extracted results
x_results = {
    (0, 6): 1.0,
    (6, 0): 1.0
}

y_results = {
    (1, 1): 1.0,
    (2, 3): 1.0,
    (3, 1): 1.0,
    (4, 0): 1.0,
    (5, 2): 1.0,
    (7, 3): 1.0,
    (8, 0): 1.0,
    (9, 3): 1.0,
    (10, 2): 1.0
}

# Create directed graphs
G_x = nx.DiGraph()
G_y = nx.DiGraph()

# Add edges to the graph for x
for (i, j), value in x_results.items():
    if value == 1.0:
        G_x.add_edge(i, j)

# Add edges to the graph for y
for (i, k), value in y_results.items():
    if value == 1.0:
        G_y.add_edge(i, k)

# Plot the x graph
plt.figure(figsize=(12, 6))
plt.subplot(121)
nx.draw(G_x, with_labels=True, node_size=700, node_color='skyblue', font_size=16, font_color='black', edge_color='gray')
plt.title('Directed Graph for x')

# Plot the y graph
plt.subplot(122)
nx.draw(G_y, with_labels=True, node_size=700, node_color='lightgreen', font_size=16, font_color='black', edge_color='gray')
plt.title('Directed Graph for y')

plt.show()
```
